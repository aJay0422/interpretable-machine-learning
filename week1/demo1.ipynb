```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt
```

# DEMO 1

```python
X = np.array([-0.289, 0.054, -1.946, -1.810, -1.611, 1.114, 1.997, 1.602, -1.063, -0.443, -1.379, -1.324, -1.274, -0.741, 0.962, 0.4866, 0.530, 0.575,
0.285, -1.622, 1.072, 0.086, -1.382, 0.213, 1.208, 0.358, 1.946, -1.115, -1.308, -0.613, -1.253, 0.615, -1.432, 0.361, 0.319, -0.269, -0.756,
1.487, 0.421, -0.599]).reshape(-1, 1)
Y = np.array([1.515, -0.514, -3.367, -1.216, 1.653, -3.682, 4.734, -1.093, 4.436,
2.459, 2.679, 3.282, 4.238, 2.984, -3.330, -2.251, -3.256, -3.454, -1.474, 1.461, -4.461, -1.603, 3.069, -2.293, -3.747, -1.929, 2.522,
3.071, 2.908, 3.788, 4.154, -3.195, 2.822, -1.969, -2.304, 1.906,
3.11, -2.620, -2.861, 3.077])
```

```python
X_seq = np.linspace(min(X), max(X), 1000).reshape(-1, 1)
```

```python
# polynomial regression(degree = 1)
degree = 1
polyreg1 = make_pipeline(PolynomialFeatures(degree=degree), LinearRegression())
polyreg1.fit(X, Y)

plt.figure()
plt.scatter(X, Y, color="blue")
plt.plot(X_seq, polyreg1.predict(X_seq), color="red")
plt.title("Polynomial regression with degree " + str(degree))
plt.show()
```

```python
# polynomial regression(degree = 3)
degree = 3
polyreg3 = make_pipeline(PolynomialFeatures(degree=degree), LinearRegression())
polyreg3.fit(X, Y)

plt.figure()
plt.scatter(X, Y, color="blue")
plt.plot(X_seq, polyreg3.predict(X_seq), color="red")
plt.title("Polynomial regression with degree " + str(degree))
plt.show()
```

```python
# polynomial regression(degree = 15)
degree = 15
polyreg15 = make_pipeline(PolynomialFeatures(degree=degree), LinearRegression())
polyreg15.fit(X, Y)

plt.figure()
plt.scatter(X, Y, color="blue")
plt.plot(X_seq, polyreg15.predict(X_seq), color="red")
plt.title("Polynomial regression with degree " + str(degree))
plt.show()
```
